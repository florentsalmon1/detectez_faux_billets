```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
import joblib  # pour sauvegarder/recharger ton mod√®le
from google.colab import drive
```

```python
drive.mount('/content/drive')
path_drive = '/content/drive/My Drive/Projet_12/data/'

# Charger les objets
model = joblib.load(path_drive + 'model_logistic.pkl')
scaler = joblib.load(path_drive + 'scaler.pkl')
```

```python
# Charger le fichier test
df_test = pd.read_csv(path_drive + 'billets_test.csv')
df_test
```

```python
# Pr√©paration des donn√©es
features = ['diagonal', 'height_left', 'height_right', 'margin_low', 'margin_up', 'length']
X_test_scaled = scaler.transform(df_test[features])

# Pr√©dictions
predictions = model.predict(X_test_scaled)
probas = model.predict_proba(X_test_scaled)

# R√©sultats
df_test['prediction'] = predictions
df_test['proba_faux'] = probas[:, 0]
df_test['proba_vrai'] = probas[:, 1]

df_test
```
